```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt


def f(u,t):
  x,y,z=u[0],u[1],u[2]


  dxdt=10*(y-x)                      #Sigma=10
  dydt=x*(28-z)-y                    #Rho=28
  dzdt=x*y-2.67*z                    #Beta=2.67
  return  [dxdt,dydt,dzdt]


t=np.arange(0,40+0.01,0.01)
u0=[0,1,0]
u=odeint(f,u0,t)


plt.title("Lorentz attractor")
plt.plot(u[:,0],u[:,2])
plt.plot(u[:,0],u[:,1])
```
